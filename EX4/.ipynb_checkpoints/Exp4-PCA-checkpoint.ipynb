## 1 ÂÆûÁé∞‰∏ªÊàêÂàÜÂàÜÊûê  

Âú®Êú¨ÈÉ®ÂàÜÂÆûÈ™å‰∏≠ÔºåÂ∞ÜÂÆûÁé∞‰∏ªÊàêÂàÜÂàÜÊûêÁÆóÊ≥ï„ÄÇ ÁÆóÊ≥ïÊ≠•È™§Â¶Ç‰∏ãÔºö

Step 1: ÂØπÊâÄÊúâÊ†∑Êú¨ËøõË°åÊ†áÂáÜÂåñ‰ΩøÂæóÊ†∑Êú¨ÁöÑÂùáÂÄº‰∏∫0ÔºåÊ†áÂáÜÂ∑Æ‰∏∫1

Step 2: ËÆ°ÁÆóÊ†∑Êú¨ÁöÑÂçèÊñπÂ∑ÆÁü©Èòµ: $\mathbf{X}\mathbf{X}^T/N$ÔºåÂÖ∂‰∏≠$N$‰∏∫Ê†∑Êú¨‰∏™Êï∞

Step 3: ÂØπÂçèÊñπÂ∑ÆÁü©Èòµ $\mathbf{X}\mathbf{X}^T/N$ ÂÅöÂ•áÂºÇÂÄºÂàÜËß£ÔºàÊàñÁâπÂæÅÂÄºÂàÜËß£Ôºâ

Step 4: ÂèñÊúÄÂ§ßÁöÑ $d$ ‰∏™ÁâπÂæÅÂÄºÊâÄÂØπÂ∫îÁöÑÁâπÂæÅÂêëÈáè $\mathbf{w}_1,\cdots,\mathbf{w}_{d}$ 

ËæìÂá∫: ÊäïÂΩ±Áü©Èòµ $\mathbf{W}=[\mathbf{w}_1,\cdots,\mathbf{w}_{d}]$ 

<div class="alert alert-block alert-warning">

**Ë¶ÅÁÇπ 1Ôºö**

Âú®‰∏ãÊñπcell‰∏≠ÔºåËØ∑**ÂÆûÁé∞''Êï∞ÊçÆÊ†áÂáÜÂåñ''ÁöÑ‰ª£Á†Å**„ÄÇ

</div>

```python
# ====================== Âú®ËøôÈáåÂ°´ÂÖ•‰ª£Á†Å =======================
def Normalize(X):
    """   
    ËæìÂÖ•
    ----------
    X : Â∞∫ÂØ∏‰∏∫ (D, N)ÁöÑÁü©ÈòµÔºåÁ¨¨iÂàó‰∏∫Á¨¨i‰∏™Ê†∑Êú¨ÔºåD‰∏∫Ê†∑Êú¨ÁöÑÁª¥Êï∞ÔºåN‰∏∫Ê†∑Êú¨ÁöÑ‰∏™Êï∞„ÄÇ 
    
    ËæìÂá∫
    -------
    X_norm : Â∞∫ÂØ∏‰∏∫ (D, N)ÁöÑÁü©Èòµ„ÄÇ 
    """  
    X1 = np.std(X, axis=1) # ÊØè‰∏ÄË°åÊ†áÂáÜÂ∑Æ
    X2 = np.mean(X, axis=1) # ÊØè‰∏ÄË°åÁöÑÂùáÂÄº
    X_norm = np.zeros(X.shape)
    N = X.shape[1]
    for i in range (N):
        X_norm[:,i] = (X[:,i] - X2)/X1
        
        
    return X_norm 
# ============================================================= 
```

<div class="alert alert-block alert-warning">

**Ë¶ÅÁÇπ 2Ôºö**

Âú®‰∏ãÊñπcell‰∏≠ÔºåËØ∑**ÂÆûÁé∞''‰∏ªÊàêÂàÜÂàÜÊûê''ÁöÑ‰ª£Á†Å**„ÄÇ

</div>

```python
# ====================== Âú®ËøôÈáåÂ°´ÂÖ•‰ª£Á†Å =======================
def PCA(X,d):
    """   
    ËæìÂÖ•
    ----------
    X : Â∞∫ÂØ∏‰∏∫ (D, N)ÁöÑÁü©ÈòµÔºåÁ¨¨iÂàó‰∏∫Á¨¨i‰∏™Ê†∑Êú¨ÔºåD‰∏∫Ê†∑Êú¨ÁöÑÁª¥Êï∞ÔºåN‰∏∫Ê†∑Êú¨ÁöÑ‰∏™Êï∞„ÄÇ 
        
    d:  ÊúüÊúõÁöÑÁª¥Êï∞
    
    ËæìÂá∫
    -------
    W : Â∞∫ÂØ∏‰∏∫ (D, d)ÁöÑÊäïÂΩ±Áü©ÈòµÔºåÁ¨¨iÂàó‰∏∫ÂçèÊñπÂ∑ÆÁü©ÈòµÁöÑÁ¨¨i‰∏™ÁâπÂæÅÂêëÈáè„ÄÇ 
    """
    D = X.shape[0]
    N = X.shape[1]
    X_T = np.zeros((d,d))
    
    # Step 1: Ê†áÂáÜÂåñÔºàË∞ÉÁî®Ë¶ÅÁÇπ1‰∏≠ÁöÑ‰ª£Á†ÅÔºâ
    
    X = Normalize(X)

    # Step 2: ËÆ°ÁÆóÂçèÊñπÂ∑ÆÁü©Èòµ
    
    X_T = np.dot(X, X.T) / N
    
    # Step 3: Â•áÂºÇÂÄºÂàÜËß£ÔºàÊü•ËØ¢numpy‰∏≠ËøõË°åÂ•áÂºÇÂÄºÂàÜËß£ÁöÑÂáΩÊï∞Ôºâ
     
    u,s,v = np.linalg.svd(X_T)

    # Step 4: ÂèñÊúÄÂ§ßÁöÑùëë‰∏™ÁâπÂæÅÂÄºÊâÄÂØπÂ∫îÁöÑÁâπÂæÅÂêëÈáèÂπ∂ÁªÑÊàêÁü©ÈòµW
    
    W = np.zeros((D,d))
    for i in range (d):
        W[:,i]= u[:,i]
    
    return W 
# ============================================================= 
```

Â¶ÇÊûúÂÆåÊàê‰∫Ü‰∏äËø∞ÂáΩÊï∞ pcaÔºå‰ª•‰∏ã‰ª£Á†ÅÂèØÁî®‰∫éÊµãËØï„ÄÇÂ¶ÇÊûúÁªìÊûú‰∏∫[-0.707107, -0.707107]ÔºåÂàôËÆ°ÁÆóÈÄöËøá„ÄÇ

```python
# Load the dataset into the variable X 
data = loadmat(os.path.join('ex4data1.mat'))
X = data['X']
X=X.T
#  Run PCA
W= PCA(X,2)

print('Á¨¨‰∏Ä‰∏™ÁâπÂæÅÂêëÈáè: W[:, 0] = [{:.6f}, {:.6f}]'.format(W[0, 0], W[1, 0]))
```

## 2 Â∞ÜPCAÂ∫îÁî®‰∫é‰ªøÁúüÊï∞ÊçÆ

Âú®Êú¨ÈÉ®ÂàÜÂÆûÈ™å‰∏≠ÔºåÂ∞ÜÂ∑≤ÂÆûÁé∞ÁöÑPCAÁÆóÊ≥ïÂ∫îÁî®‰∫éÊï∞ÊçÆÈõÜ1ÔºåËØ•Êï∞ÊçÆÈõÜ‰∏≠ÁöÑÊ†∑Êú¨Áª¥Êï∞‰∏∫2ÔºåÂõ†Ê≠§ÈôçÁª¥ÁªìÊùüÂêéÔºåÂèØÈÄöËøáÂèØËßÜÂåñËßÇÂØüÈôçÁª¥ÂâçÂêéÁöÑÁªìÊûú„ÄÇ

```python
#  ÂèØËßÜÂåñÂéüÂßãÊï∞ÊçÆÈõÜ
plt.plot(X[0, :], X[1, :], 'bo', ms=10, mec='k', mew=1)
plt.axis([0.5, 6.5, 2, 8])
plt.gca().set_aspect('equal')
plt.grid(False)
```

```python
#  ÂèØËßÜÂåñÊ†áÂáÜÂåñ‰πãÂêéÁöÑÊï∞ÊçÆÈõÜ
X_norm=Normalize(X)
plt.plot(X_norm[0, :], X_norm[1, :], 'bo', ms=10, mec='k', mew=1)
plt.axis([-3, 2.75, -3, 2.75])
plt.gca().set_aspect('equal')
plt.grid(False)
```

### 2.1 Âà©Áî®PCAÈôçÁª¥
Âà©Áî®‰∏äËø∞PCA‰ª£Á†ÅÂèØÂ∞Ü‰ªøÁúüÊï∞ÊçÆÈôçËá≥1Áª¥„ÄÇ

<a id="projectData"></a>


<div class="alert alert-block alert-warning">

**Ë¶ÅÁÇπ 3Ôºö**

Âú®‰∏ãÊñπcell‰∏≠ÔºåËØ∑**ÂÆûÁé∞''ProjectData''ÁöÑ‰ª£Á†Å**ÔºåËæìÂá∫ÈôçÁª¥ÂêéÁöÑÊï∞ÊçÆ„ÄÇ

</div>

```python
# ====================== Âú®ËøôÈáåÂ°´ÂÖ•‰ª£Á†Å =======================
def ProjectData(X, W):
    """   
    ËæìÂÖ•
    ----------
    X : Â∞∫ÂØ∏‰∏∫ (D, N)ÁöÑÁü©ÈòµÔºåÁ¨¨iÂàó‰∏∫Á¨¨i‰∏™Ê†∑Êú¨ÔºåD‰∏∫Ê†∑Êú¨ÁöÑÁª¥Êï∞ÔºåN‰∏∫Ê†∑Êú¨ÁöÑ‰∏™Êï∞„ÄÇ 
    W : Â∞∫ÂØ∏‰∏∫ (D, d)ÁöÑÊäïÂΩ±Áü©ÈòµÔºåÁ¨¨iÂàó‰∏∫ÂçèÊñπÂ∑ÆÁü©ÈòµÁöÑÁ¨¨i‰∏™ÁâπÂæÅÂêëÈáè„ÄÇ   
    
    ËæìÂá∫
    -------
    Z : Â∞∫ÂØ∏‰∏∫ (d, N)ÁöÑÁü©ÈòµÔºåÁ¨¨iÂàó‰∏∫Á¨¨i‰∏™ÈôçÁª¥ÂêéÁöÑÊï∞ÊçÆ„ÄÇ 
    """

    # Step 1: Ê†áÂáÜÂåñÔºàË∞ÉÁî®Ë¶ÅÁÇπ1‰∏≠ÁöÑ‰ª£Á†ÅÔºâ
    X = Normalize(X)
    
    # Step 2: ËÆ°ÁÆóÈôçÁª¥ÂêéÁöÑÊï∞ÊçÆ
    Z = np.dot(W.T,X)

    return Z
# ============================================================= 
```

Â¶ÇÊûúÂÆåÊàê‰∫Ü‰∏äËø∞ÂáΩÊï∞ ProjectDataÔºå‰ª•‰∏ã‰ª£Á†ÅÂèØÁî®‰∫éÊµãËØï„ÄÇÂ¶ÇÊûúÁªìÊûú‰∏∫ 1.496313ÔºåÂàôËÆ°ÁÆóÈÄöËøá„ÄÇ

```python
#  Â∞ÜÊï∞ÊçÆÈôçËá≥‰∏ÄÁª¥ 
W = PCA(X,1)
Z = ProjectData(X, W)
print('Á¨¨‰∏Ä‰∏™Ê†∑Êú¨ÈôçÁª¥ÂêéÁöÑÊï∞ÊçÆ: {:.6f}'.format(Z[0, 0]))
```

### 2.2 ÈÄöËøáÈôçÁª¥ÂêéÁöÑÊï∞ÊçÆÈáçÊûÑÂéüÂßãÊï∞ÊçÆ

Âà©Áî®ÈôçÁª¥ÂêéÁöÑÊï∞ÊçÆÁü©Èòµ$\mathbf{Z}$ÔºåÂèØËøë‰ººÈáçÊûÑÂéüÂßãÊï∞ÊçÆ„ÄÇ
<a id="recoverData"></a>

<div class="alert alert-block alert-warning">

**Ë¶ÅÁÇπ 4Ôºö**

Âú®‰∏ãÊñπcell‰∏≠ÔºåËØ∑**ÂÆûÁé∞''RecoverData''ÁöÑ‰ª£Á†Å**ÔºåËæìÂá∫ÈáçÊûÑÂêéÁöÑÊï∞ÊçÆ„ÄÇ

</div>

```python
# ====================== Âú®ËøôÈáåÂ°´ÂÖ•‰ª£Á†Å =======================
def RecoverData(Z, W):
    """   
    ËæìÂÖ•
    ----------
    Z : Â∞∫ÂØ∏‰∏∫ (d, N)ÁöÑÁü©ÈòµÔºåÁ¨¨iÂàó‰∏∫Á¨¨i‰∏™ÈôçÁª¥ÂêéÁöÑÊï∞ÊçÆ„ÄÇ 
    W : Â∞∫ÂØ∏‰∏∫ (D, d)ÁöÑÊäïÂΩ±Áü©ÈòµÔºåÁ¨¨iÂàó‰∏∫ÂçèÊñπÂ∑ÆÁü©ÈòµÁöÑÁ¨¨i‰∏™ÁâπÂæÅÂêëÈáè„ÄÇ   
    
    ËæìÂá∫
    -------
    X_rec : Â∞∫ÂØ∏‰∏∫ (D, N)ÁöÑÁü©ÈòµÔºåÁ¨¨iÂàó‰∏∫Á¨¨i‰∏™ÈáçÊûÑÂêéÁöÑÊï∞ÊçÆ„ÄÇ 
    """
    X_rec = np.dot(W, Z)
    return X_rec
# ============================================================= 
```

Â¶ÇÊûúÂÆåÊàê‰∫Ü‰∏äËø∞ÂáΩÊï∞ RecoverDataÔºå‰ª•‰∏ã‰ª£Á†ÅÂèØÁî®‰∫éÊµãËØï„ÄÇÂ¶ÇÊûúÁªìÊûú‰∏∫[-1.058053, -1.058053]ÔºåÂàôËÆ°ÁÆóÈÄöËøá„ÄÇ

```python
#  ÈáçÊûÑÂéüÂßãÊï∞ÊçÆ
X_rec = RecoverData(Z, W)
# print('Á¨¨‰∏Ä‰∏™ÈáçÊûÑÂêéÁöÑÊï∞ÊçÆ: {:.6f}'.format(X_rec[:, 0]))
print('Á¨¨‰∏Ä‰∏™ÈáçÊûÑÂêéÁöÑÊï∞ÊçÆ: X_rec[:, 0] = [{:.6f}, {:.6f}]'.format(X_rec[0, 0], X_rec[1, 0]))
```

```python
#ÂèØËßÜÂåñÈáçÊûÑÂâçÂêéÁöÑÁªìÊûú
X_norm=Normalize(X) 

#  Plot the normalized dataset (returned from featureNormalize)
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(X_norm[0, :], X_norm[1, :], 'bo', ms=8, mec='b', mew=0.5)
ax.set_aspect('equal')
ax.grid(False)
plt.axis([-3, 2.75, -3, 2.75])

# Draw lines connecting the projected points to the original points
ax.plot(X_rec[0, :], X_rec[1, :], 'ro', mec='r', mew=2, mfc='none')
for xnorm, xrec in zip(X_norm.T, X_rec.T):
    ax.plot([xnorm[0], xrec[0]], [xnorm[1], xrec[1]], '--k', lw=1)
```

## 3 Â∞ÜPCAÂ∫îÁî®‰∫é‰∫∫ËÑ∏ÈáçÊûÑ

Âú®Êú¨ÈÉ®ÂàÜÂÆûÈ™å‰∏≠ÔºåÂ∞ÜÂ∑≤ÂÆûÁé∞ÁöÑPCAÁÆóÊ≥ïÂ∫îÁî®‰∫éÊï∞ÊçÆÈõÜ2Ôºå‰∏∫LFW‰∫∫ËÑ∏Êï∞ÊçÆÈõÜÂêàÁöÑÂ≠êÈõÜ„ÄÇ

```python
#  Load Face dataset
data = loadmat(os.path.join('ex4data2.mat'))
X = data['X']
X=X.T
X = X[:,:100]
```

```python
#ÂÆö‰πâÊòæÁ§∫‰∫∫ËÑ∏ÂõæÂÉèÂáΩÊï∞
def displayImage(X):
    """   
    ËæìÂÖ•
    ----------
    X : Â∞∫ÂØ∏‰∏∫ (D, N)ÁöÑÁü©ÈòµÔºåÁ¨¨iÂàó‰∏∫Á¨¨i‰∏™Ê†∑Êú¨ÔºåD‰∏∫Ê†∑Êú¨ÁöÑÁª¥Êï∞ÔºåN‰∏∫Ê†∑Êú¨ÁöÑ‰∏™Êï∞„ÄÇ   
    """
    # Compute number of items to display
    N=X.shape[1]
    display_rows = int(np.floor(np.sqrt(N)))
    display_cols = int(np.ceil(N / display_rows))
    
    figsize=(N,N)
    fig, ax_array = plt.subplots(display_rows, display_cols)

    ax_array = [ax_array] if N == 1 else ax_array.ravel()
    for i, ax in enumerate(ax_array):
        ax.imshow(X[:,i].reshape(32, 32, order='F'), cmap='gray')
        ax.axis('off')
```

### 3.1 Âà©Áî®PCAÂØπ‰∫∫ËÑ∏ÂõæÂÉèÈôçÁª¥

<a id="projectData"></a>

<div class="alert alert-block alert-warning">

**Ë¶ÅÁÇπ 5Ôºö**

Âà©Áî®PCA‰ª£Á†ÅÂ∞Ü‰∫∫ËÑ∏Êï∞ÊçÆÈôçËá≥64Áª¥ÔºåÂπ∂ÊòæÁ§∫Ââç16‰∏™ÁâπÂæÅÂêëÈáè„ÄÇ

</div>

```python
# ====================== Âú®ËøôÈáåÂ°´ÂÖ•‰ª£Á†Å =======================

#Step 1: Âà©Áî®PCAÂØπÂéüÂßãÊï∞ÊçÆÈôçÁª¥
D,N = X.shape

W = PCA(X,64)

Z = ProjectData(X,W)

#Step 2: Ë∞ÉÁî®ÂâçÊñá‰∏≠ÁöÑdisplayImageÂáΩÊï∞ÊòæÁ§∫Ââç16‰∏™ÁâπÂæÅÂêëÈáèÊâÄÂØπÂ∫îÁöÑÂõæÂÉè

X = W[: , : 16]

displayImage(X)

# ============================================================= 
```

### 3.2 ÈÄöËøá‰∏ªÊàêÂàÜÈáçÊûÑÂéüÂßã‰∫∫ËÑ∏Êï∞ÊçÆ

<div class="alert alert-block alert-warning">

**Ë¶ÅÁÇπ 6Ôºö**

Âà©Áî®ÈôçÁª¥ÂêéÁöÑÊï∞ÊçÆÈáçÊûÑÂéüÂßã‰∫∫ËÑ∏Êï∞ÊçÆÔºåÂπ∂ÊòæÁ§∫Ââç16‰∏™ÈáçÊûÑÂâçÂêéÁöÑ‰∫∫ËÑ∏Êï∞ÊçÆ„ÄÇ

</div>

```python
# ====================== Âú®ËøôÈáåÂ°´ÂÖ•‰ª£Á†Å =======================

#Step 1: Âà©Áî®RecoverDataÂáΩÊï∞ÂæóÂà∞ÈáçÊûÑÊï∞ÊçÆ 

X_rec = RecoverData(Z, W)

#Step 2: Âà©Áî®displayImageÂáΩÊï∞ÊòæÁ§∫Ââç16‰∏™ÂéüÂßãÂõæÂÉè  

displayImage(X[: , 0:16])

plt.gcf().suptitle('Original faces')


#Step 3: Âà©Áî®displayImageÂáΩÊï∞ÊòæÁ§∫Ââç16‰∏™ÈáçÊûÑÂõæÂÉè 

displayImage(X_rec[: , 0:16])
plt.gcf().suptitle('Recovered faces')
 
# =============================================================  
```
