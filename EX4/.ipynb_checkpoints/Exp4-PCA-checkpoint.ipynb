## 1 实现主成分分析  

在本部分实验中，将实现主成分分析算法。 算法步骤如下：

Step 1: 对所有样本进行标准化使得样本的均值为0，标准差为1

Step 2: 计算样本的协方差矩阵: $\mathbf{X}\mathbf{X}^T/N$，其中$N$为样本个数

Step 3: 对协方差矩阵 $\mathbf{X}\mathbf{X}^T/N$ 做奇异值分解（或特征值分解）

Step 4: 取最大的 $d$ 个特征值所对应的特征向量 $\mathbf{w}_1,\cdots,\mathbf{w}_{d}$ 

输出: 投影矩阵 $\mathbf{W}=[\mathbf{w}_1,\cdots,\mathbf{w}_{d}]$ 

<div class="alert alert-block alert-warning">

**要点 1：**

在下方cell中，请**实现''数据标准化''的代码**。

</div>

```python
# ====================== 在这里填入代码 =======================
def Normalize(X):
    """   
    输入
    ----------
    X : 尺寸为 (D, N)的矩阵，第i列为第i个样本，D为样本的维数，N为样本的个数。 
    
    输出
    -------
    X_norm : 尺寸为 (D, N)的矩阵。 
    """  
    X1 = np.std(X, axis=1) # 每一行标准差
    X2 = np.mean(X, axis=1) # 每一行的均值
    X_norm = np.zeros(X.shape)
    N = X.shape[1]
    for i in range (N):
        X_norm[:,i] = (X[:,i] - X2)/X1
        
        
    return X_norm 
# ============================================================= 
```

<div class="alert alert-block alert-warning">

**要点 2：**

在下方cell中，请**实现''主成分分析''的代码**。

</div>

```python
# ====================== 在这里填入代码 =======================
def PCA(X,d):
    """   
    输入
    ----------
    X : 尺寸为 (D, N)的矩阵，第i列为第i个样本，D为样本的维数，N为样本的个数。 
        
    d:  期望的维数
    
    输出
    -------
    W : 尺寸为 (D, d)的投影矩阵，第i列为协方差矩阵的第i个特征向量。 
    """
    D = X.shape[0]
    N = X.shape[1]
    X_T = np.zeros((d,d))
    
    # Step 1: 标准化（调用要点1中的代码）
    
    X = Normalize(X)

    # Step 2: 计算协方差矩阵
    
    X_T = np.dot(X, X.T) / N
    
    # Step 3: 奇异值分解（查询numpy中进行奇异值分解的函数）
     
    u,s,v = np.linalg.svd(X_T)

    # Step 4: 取最大的𝑑个特征值所对应的特征向量并组成矩阵W
    
    W = np.zeros((D,d))
    for i in range (d):
        W[:,i]= u[:,i]
    
    return W 
# ============================================================= 
```

如果完成了上述函数 pca，以下代码可用于测试。如果结果为[-0.707107, -0.707107]，则计算通过。

```python
# Load the dataset into the variable X 
data = loadmat(os.path.join('ex4data1.mat'))
X = data['X']
X=X.T
#  Run PCA
W= PCA(X,2)

print('第一个特征向量: W[:, 0] = [{:.6f}, {:.6f}]'.format(W[0, 0], W[1, 0]))
```

### 2.1 利用PCA降维
利用上述PCA代码可将仿真数据降至1维。

<a id="projectData"></a>


<div class="alert alert-block alert-warning">

**要点 3：**

在下方cell中，请**实现''ProjectData''的代码**，输出降维后的数据。

</div>

```python
# ====================== 在这里填入代码 =======================
def ProjectData(X, W):
    """   
    输入
    ----------
    X : 尺寸为 (D, N)的矩阵，第i列为第i个样本，D为样本的维数，N为样本的个数。 
    W : 尺寸为 (D, d)的投影矩阵，第i列为协方差矩阵的第i个特征向量。   
    
    输出
    -------
    Z : 尺寸为 (d, N)的矩阵，第i列为第i个降维后的数据。 
    """

    # Step 1: 标准化（调用要点1中的代码）
    X = Normalize(X)
    
    # Step 2: 计算降维后的数据
    Z = np.dot(W.T,X)

    return Z
# ============================================================= 
```

如果完成了上述函数 ProjectData，以下代码可用于测试。如果结果为 1.496313，则计算通过。

```python
#  将数据降至一维 
W = PCA(X,1)
Z = ProjectData(X, W)
print('第一个样本降维后的数据: {:.6f}'.format(Z[0, 0]))
```

<a id="section5"></a>
### 2.2 通过降维后的数据重构原始数据

利用降维后的数据矩阵$\mathbf{Z}$，可近似重构原始数据。
<a id="recoverData"></a>

<div class="alert alert-block alert-warning">

**要点 4：**

在下方cell中，请**实现''RecoverData''的代码**，输出重构后的数据。

</div>

```python
# ====================== 在这里填入代码 =======================
def RecoverData(Z, W):
    """   
    输入
    ----------
    Z : 尺寸为 (d, N)的矩阵，第i列为第i个降维后的数据。 
    W : 尺寸为 (D, d)的投影矩阵，第i列为协方差矩阵的第i个特征向量。   
    
    输出
    -------
    X_rec : 尺寸为 (D, N)的矩阵，第i列为第i个重构后的数据。 
    """
    X_rec = np.dot(W, Z)
    return X_rec
# ============================================================= 
```

如果完成了上述函数 RecoverData，以下代码可用于测试。如果结果为[-1.058053, -1.058053]，则计算通过。

```python
#  重构原始数据
X_rec = RecoverData(Z, W)
# print('第一个重构后的数据: {:.6f}'.format(X_rec[:, 0]))
print('第一个重构后的数据: X_rec[:, 0] = [{:.6f}, {:.6f}]'.format(X_rec[0, 0], X_rec[1, 0]))
```

```python
#可视化重构前后的结果
X_norm=Normalize(X) 

#  Plot the normalized dataset (returned from featureNormalize)
fig, ax = plt.subplots(figsize=(5, 5))
ax.plot(X_norm[0, :], X_norm[1, :], 'bo', ms=8, mec='b', mew=0.5)
ax.set_aspect('equal')
ax.grid(False)
plt.axis([-3, 2.75, -3, 2.75])

# Draw lines connecting the projected points to the original points
ax.plot(X_rec[0, :], X_rec[1, :], 'ro', mec='r', mew=2, mfc='none')
for xnorm, xrec in zip(X_norm.T, X_rec.T):
    ax.plot([xnorm[0], xrec[0]], [xnorm[1], xrec[1]], '--k', lw=1)
```

## 3 将PCA应用于人脸重构

在本部分实验中，将已实现的PCA算法应用于数据集2，为LFW人脸数据集合的子集。

```python
#  Load Face dataset
data = loadmat(os.path.join('ex4data2.mat'))
X = data['X']
X=X.T
X = X[:,:100]
```

```python
#定义显示人脸图像函数
def displayImage(X):
    """   
    输入
    ----------
    X : 尺寸为 (D, N)的矩阵，第i列为第i个样本，D为样本的维数，N为样本的个数。   
    """
    # Compute number of items to display
    N=X.shape[1]
    display_rows = int(np.floor(np.sqrt(N)))
    display_cols = int(np.ceil(N / display_rows))
    
    figsize=(N,N)
    fig, ax_array = plt.subplots(display_rows, display_cols)

    ax_array = [ax_array] if N == 1 else ax_array.ravel()
    for i, ax in enumerate(ax_array):
        ax.imshow(X[:,i].reshape(32, 32, order='F'), cmap='gray')
        ax.axis('off')
```
